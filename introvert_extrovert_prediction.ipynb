```python
df = pd.read_csv('train.csv')
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer, KNNImputer
```

```python
df.info()
```

```python
# df = df.dropna()
```

```python
def boolToint(a):
    if (pd.isna(a) == False):    
        if(a.lower() == 'no'):
            return 0
        else:
            return 1
```

```python
df['Stage_fear'] = df['Stage_fear'].apply(boolToint)
```

```python
df['Drained_after_socializing'] = df['Drained_after_socializing'].apply(boolToint)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.duplicated().value_counts()
```

```python
df.info()
```

```python
df = df.drop(columns=['id'])
```

```python
df.sample(5, random_state = 10)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
scaler = MinMaxScaler()
df.iloc[:, 0:7] = scaler.fit_transform(df.iloc[: , 0:7])
```

```python
df.sample(5, random_state= 10)
```

```python
X= df.iloc[:, 0:7]
y= df.iloc[:, 7]
```

```python
# imp = IterativeImputer(max_iter=10)
imp = KNNImputer(n_neighbors=5)
imp.set_output(transform = "pandas")
X = imp.fit_transform(X)
```

```python
X.sample(5, random_state=10)
```

```python
# df2.sample(5, random_state = 10)
```

```python
y
```

```python
from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()
```

```python
y = lb.fit_transform(y)
```

```python
lb.classes_
```

```python
y
```

```python
np.unique(y)
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=4)
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(
    n_estimators=1000,
    max_depth=14,
    criterion='entropy',
    min_samples_split=10
)
```

```python
# from imblearn.ensemble import  BalancedRandomForestClassifier
# rf = BalancedRandomForestClassifier(
#     n_estimators=1000,
#     max_depth=14,
#     criterion='entropy',
#     min_samples_split=10,
#     random_state=5
#     )
```

```python
rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
pred = rf.predict(X_test)
```

```python
print(classification_report(y_test, pred))
```

```python
pred2 = rf.predict(X_train)
print(classification_report(y_train, pred2))
```
